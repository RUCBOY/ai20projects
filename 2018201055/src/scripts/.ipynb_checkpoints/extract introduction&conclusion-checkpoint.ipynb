```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
#发现sciencedirect上的introduction和conclusion两部分在登陆后才动态加载，需要引入selenium模拟浏览器爬虫
```

```python
urllist=["https://www.sciencedirect.com/search?articleTypes=FLA&navigation=true&qs=computer%20science&show=100&sortBy=relevance&years=2021&lastSelectedFacet=years"]
#Elsevier Sciencedirect数据库的url，关键词搜索“computer science”,选择“研究文章”，按年查找，按相关性排序，每页显示100条
#构建urllist列表，第一个元素为每年搜索结果第一页的url，每次手动勾选年份，手动在urllist列表中修改第一个元素

for i in range(1,10):
    url="https://www.sciencedirect.com/search?articleTypes=FLA&navigation=true&qs=computer%%20science&show=100&sortBy=relevance&years=2021&offset=%d"%(i)+"00"
    urllist.append(url)
#将每年的2-10页搜索结果的url添加到urllist列表中

print(len(urllist))
urllist
```

```python
#遍历列表中的每个url，爬取每一页中每篇论文的二层网页链接，并保存在列表linklist中，共1000条论文链接
linklist=[]
for url in urllist:
    res=requests.get(url,headers={"user-agent":"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36"})
    print(res)
    time.sleep(1)
    soup=BeautifulSoup(res.content, 'html.parser')
    soup2=soup.find_all('li',{"class":"ResultItem"})
    for i in soup2:
        s=i.find('h2').find('a')['href']
        link='https://www.sciencedirect.com'+s
        linklist.append(link)

print(len(linklist))
linklist
```

```python
#先手动在一个chrome浏览器中登陆sciencedirect，再使用selenium接管chrome，以便得到动态加载后的数据
chrome_options = Options()
chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
chrome_driver = "C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe"
```

```python
#遍历linklist列表中的每一项，连接到每篇论文的网页，爬取Introduction并保存到txt文档中,若查找不到则跳过
intro_num=1 #每次修改年份后手动更改intro_num的值
for link in linklist[:]:
    driver = webdriver.Chrome(chrome_driver, options=chrome_options)#启动已手动调试好的chrome浏览器
    driver.get(link)
    time.sleep(2)
    bs=BeautifulSoup(driver.page_source,'lxml')
    try:
        bs2=bs.find('div',{"id":"body"}).find_all('h2')
        for h2 in bs2:
            h=h2.text.find("Introduction")
            if h !=-1:
                intros=h2.next_siblings
                for i in intros:
                    intro=i.text+'\n'
                    with open('introduction/%d'%(intro_num)+'.txt','a',encoding='utf-8') as f:
                        f.write(intro)
                f.close()
                print('downloaded%d'%(intro_num))
    except:
        pass
    intro_num+=1
```

```python
#遍历linklist列表中的每一项，连接到每篇论文的网页，爬取Conclusion并保存到txt文档中,若查找不到则跳过
conc_num=2001 #每次修改年份后手动更改conc_num的值
for link in linklist[:]:
    driver = webdriver.Chrome(chrome_driver, options=chrome_options)#启动已手动调试好的chrome浏览器
    driver.get(link)
    time.sleep(2)
    bs=BeautifulSoup(driver.page_source,'lxml')
    try:
        bs2=bs.find('div',{"id":"body"}).find_all('h2')
        for h2 in bs2:
            h=h2.text.find("Conclusion")
            if h !=-1:
                concs=h2.next_siblings
                for i in concs:
                    conc=i.text+'\n'
                    with open('conclusion/%d'%(conc_num)+'.txt','a',encoding='utf-8') as f:
                        f.write(conc)
                f.close()
                print('downloaded%d'%(conc_num))
    except:
        pass
    conc_num+=1
```
