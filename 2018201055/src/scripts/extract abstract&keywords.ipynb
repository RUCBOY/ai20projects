```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
urllist=["https://www.sciencedirect.com/search?articleTypes=FLA&navigation=true&qs=computer%20science&show=100&sortBy=relevance&years=2017&lastSelectedFacet=years"]
#Elsevier Sciencedirect数据库的url，关键词搜索“computer science”,选择“研究文章”，按年查找，按相关性排序，每页显示100条
#构建urllist列表，第一个元素为每年搜索结果第一页的url，每次手动勾选年份，手动在urllist列表中修改第一个元素

for i in range(1,10):
    url="https://www.sciencedirect.com/search?articleTypes=FLA&navigation=true&qs=computer%%20science&show=100&sortBy=relevance&years=2017&offset=%d"%(i)+"00"
    urllist.append(url)
#将每年的2-10页搜索结果的url添加到urllist列表中

print(len(urllist))
urllist
```

```python
#遍历列表中的每个url，爬取每一页中每篇论文的二层网页链接，并保存在列表linklist中，共1000条论文链接
linklist=[]
for url in urllist:
    res=requests.get(url,headers={"user-agent":"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36"})
    print(res)
    time.sleep(1)
    soup=BeautifulSoup(res.content, 'html.parser')
    soup2=soup.find_all('li',{"class":"ResultItem"})
    for i in soup2:
        s=i.find('h2').find('a')['href']
        link='https://www.sciencedirect.com'+s
        linklist.append(link)

print(len(linklist))
linklist
```

```python
#遍历linklist列表中的每一项，连接到每篇论文的网页，爬取Abstract并保存到txt文档中,若查找不到则跳过
abst_num=4001 #每次修改年份后手动更改abst_num的值
for link in linklist:
    res2=requests.get(link,headers={"user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36"})
    print(res2)
    res2.encoding='utf-8'
    soup3=BeautifulSoup(res2.text, 'html.parser')    
    try:
        soup4=soup3.find('div',{"class":"abstract author"}).find_all('p')
        for i in soup4:
            abst=i.text+'\n'
            with open('abstract/%d'%(abst_num)+'.txt','a',encoding='utf-8') as f:
                f.write(abst)
        print('downloaded%d'%(abst_num))
        f.close()
    except:
        pass
    abst_num+=1
```

```python
#遍历linklist列表中的每一项，连接到每篇论文的网页，爬取Keywords并保存到txt文档中,若查找不到则跳过
key_num=4389 #每次修改年份后手动更改key_num的值
for link in linklist[388:]:
    res2=requests.get(link,headers={"user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36"})
    print(res2)
    res2.encoding='utf-8'
    soup3=BeautifulSoup(res2.text, 'html.parser')
    try:
        soup5=soup3.find_all('div',{"class":"keyword"})
        for i in soup5[0:-1]:
            key=i.find('span').text+';'
            with open('keywords/%d'%(key_num)+'.txt','a',encoding='utf-8') as f:
                f.write(key)  
        key=soup5[-1].find('span').text
        with open('keywords/%d'%(key_num)+'.txt','a',encoding='utf-8') as f:
            f.write(key)    
        print('downloaded%d'%(key_num))
        f.close()
    except:
        pass
    key_num+=1
```
