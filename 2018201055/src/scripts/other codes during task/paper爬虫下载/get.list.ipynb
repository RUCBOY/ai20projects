```python
import requests
from bs4 import BeautifulSoup
```

```python
#连接DBIP数据库的网页
url="https://dblp.uni-trier.de/db/conf/coling/coling2018.html"
res=requests.get(url)
print(res)
```

```python
soup=BeautifulSoup(res.content, 'html.parser')
soup2=soup.find_all('li',{"class":"entry inproceedings"})
```

```python
#提取和保存论文的DOI链接
fp=open('paperlink2.txt','w')
for i in soup2:
    s = i.find('a')['href']+'\n'
    sci_link=s
    print(sci_link)
    fp.write(sci_link)
fp.close()
```

```python
#提取和保存论文的Title
for i in soup2:
    soup3=i.find_all('span',{"class":"title"})
    for i in soup3:
        s=i.string
        s=i.text+'\n'
        print(s)
        fpname=open('papername2.txt','a',encoding='utf-8')
        fpname.write(s)
fpname.close()
```
